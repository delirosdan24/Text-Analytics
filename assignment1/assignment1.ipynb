```python
# Name: Muhammad Adli bin Rosdan
# ID: SW01083749
# Name: Muhammad Irfan Azraei bin Izhar Kamil
# ID: SW01083594

import requests
from bs4 import BeautifulSoup
import csv
import time

# Purpose: This function handles the scraping logic across 5 pages [cite: 8, 15]
def scrape_reviews(domain):
    all_reviews = []
    # Standard headers to prevent basic security blocks
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36'
    }

    # Requirement: Loop through exactly 5 pages [cite: 8]
    for page_num in range(1, 6):
        print(f"Scraping page {page_num}...")
        url = f"https://www.trustpilot.com/review/{domain}?page={page_num}"
        
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Finding all review containers using the 'article' tag
        review_elements = soup.find_all('article')

        for item in review_elements:
            try:
                # a. Extract Reviewer Name [cite: 11]
                name = item.find('span', {'data-consumer-name-typography': True}).text.strip()
                
                # b. Extract Review Date (Standardized format) [cite: 12]
                date = item.find('time')['datetime'].split('T')[0]
                
                # c. Extract Review Content (Combining Title and Body for complete data) [cite: 13]
                title = item.find('h2', {'data-service-review-title-typography': True}).text.strip()
                body_tag = item.find('p', {'data-review-content-typography': True})
                body = body_tag.text.strip() if body_tag else ""
                
                # Merging them into one field to satisfy the assignment requirements [cite: 13]
                combined_content = f"{title}: {body}".strip(": ")

                all_reviews.append({
                    'Reviewer name': name,
                    'Review date': date,
                    'Review content': combined_content
                })
            except Exception:
                # Skip if there is an error with a specific review
                continue
        
        # Pause to mimic human browsing behavior
        time.sleep(2)
        
    return all_reviews

# Purpose: This function saves the data to a CSV file using DictWriter [cite: 14, 15]
def save_to_csv(data_list, filename):
    # This logic matches your Lab 3 implementation
    with open(filename, 'w', encoding='utf-8', newline='') as csvfile:
        fieldnames = ['Reviewer name', 'Review date', 'Review content']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        writer.writerows(data_list)
            
    print(f"Data successfully saved to {filename}")

# --- Execution ---
target_domain = "www.samsung.com" # Reliable real-world platform [cite: 7]
scraped_data = scrape_reviews(target_domain)
save_to_csv(scraped_data, 'reviews.csv')
```
