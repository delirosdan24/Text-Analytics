Lexicon-based method (VADER):

The main strength of the lexicon-based method is its simplicity, speed, and ease of implementation, since it does not require model training and can classify sentiment directly using a predefined sentiment dictionary. In your result, VADER performed reasonably well for the positive class, with a high positive F1-score of 0.89 and overall accuracy of 0.8051, which shows that it can detect clearly positive reviews quite effectively. However, its weaknesses are obvious: it performed much worse for negative reviews and almost completely failed on neutral reviews, with a neutral F1-score of only 0.06. This shows that VADER struggles with subtle, mixed, or context-dependent language because it relies only on fixed word sentiment scores rather than learning patterns from the actual dataset.

.

Bag-of-Words (BoW) Machine Learning Based method:

The strength of the Bag-of-Words method is that it converts text into a structured numerical form that machine learning models can learn from, making it more effective than a basic lexicon approach for dataset-specific sentiment classification. In your results, BoW combined with LinearSVC achieved a strong accuracy of 0.8584, outperforming both VADER and BoW with Naive Bayes, while also producing solid performance for the positive and negative classes. BoW with Naive Bayes was also relatively stable, especially compared with TF-IDF + Naive Bayes. However, the weakness of BoW is that it only counts word occurrences and ignores deeper context, meaning, and word importance across documents. Because of that, both BoW models still struggled badly with the neutral class, especially LinearSVC, where neutral F1 dropped to 0.18. So BoW is clearly stronger than the lexicon method, but it still has trouble separating classes that use overlapping language.

.

TF-IDF Machine Learning Based method:

The strength of the TF-IDF method is that it improves text representation by giving more importance to informative words and reducing the influence of very common words, which often helps machine learning models make better decisions. In your results, TF-IDF combined with LinearSVC achieved the highest accuracy of all models at 0.8638, making it the strongest overall approach. It also slightly improved negative classification compared with BoW + LinearSVC, while maintaining very strong performance for the positive class. However, TF-IDF is not perfect. When paired with Naive Bayes, it became less balanced, with the negative F1-score dropping to 0.41 and the neutral F1-score collapsing to 0.00, which is terrible. This means TF-IDF is powerful, but its effectiveness depends heavily on the classifier used. Like the other methods, it still failed to handle the neutral class properly, showing that class imbalance and overlapping review language remain major weaknesses.

.

Conclusion:

Among the three methods, TF-IDF is the best overall method, because when combined with LinearSVC, it produced the highest accuracy of 86.38% and the strongest overall classification performance. Although BoW also performed well, especially with LinearSVC, TF-IDF gave a slight improvement and therefore stands as the best feature extraction method in this comparison. The lexicon-based method was the weakest because it could not learn from the dataset and performed poorly on negative and neutral reviews. So the best overall choice from your experiment is TF-IDF with LinearSVC, even though all methods still showed a clear weakness in classifying neutral sentiment.
